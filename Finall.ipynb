```python
# TODO: ุชุตุงูุฑ ุจโฺฉูุช (LR) ุฑุง ุจุง ฺฉูฺฺฉ ฺฉุฑุฏู ุชุตุงูุฑ ุจุงฺฉูุช (HR) ุจุณุงุฒุฏ.

from sklearn.model_selection import train_test_split

def make_lr_from_hr(hr_images, scale=2, down_method='area', antialias=True, as_numpy=False):

    hr_tf = tf.convert_to_tensor(hr_images, dtype=tf.float32)
    h = tf.shape(hr_tf)[1]
    w = tf.shape(hr_tf)[2]
    lr_tf = tf.image.resize(hr_tf,
                            (h // scale, w // scale),
                            method=down_method,
                            antialias=antialias)
    return lr_tf.numpy() if as_numpy else lr_tf


SCALE = 2
x_train_lr = make_lr_from_hr(x_train, scale=SCALE, down_method='area', antialias=True, as_numpy=False)
x_test_lr  = make_lr_from_hr(x_test,  scale=SCALE, down_method='area', antialias=True, as_numpy=False)

print("LR shapes:",
      x_train_lr.shape, x_test_lr.shape)


idx = np.arange(len(x_train))
train_idx, val_idx = train_test_split(idx, test_size=0.1, random_state=42, shuffle=True)

x_train_hr = tf.gather(x_train, train_idx)
x_val_hr   = tf.gather(x_train, val_idx)
x_train_lr = tf.gather(x_train_lr, train_idx)
x_val_lr   = tf.gather(x_train_lr, val_idx)

print("After split:",
      "x_train_hr", x_train_hr.shape,
      "x_train_lr", x_train_lr.shape,
      "x_val_hr",   x_val_hr.shape,
      "x_val_lr",   x_val_lr.shape)

# ุงฺฏุฑ ุจุฑุง Keras ุจู numpy ูุงุฒ ุฏุงุฑ:
x_train_hr_np = x_train_hr.numpy()
x_train_lr_np = x_train_lr.numpy()
x_val_hr_np   = x_val_hr.numpy()
x_val_lr_np   = x_val_lr.numpy()
x_test_hr_np  = x_test.astype('float32')  # HR ุชุณุช ููุงู x_test ุงุณุช
x_test_lr_np  = x_test_lr.numpy()

```

```python
# TODO: ฺฉ ุชุงุจุน ุจุฑุง ููุงุด ฺฉ ุชุตูุฑ LR ู HR ุฏุฑ ฺฉูุงุฑ ูู ุจููุณุฏ.

def _to_float01(img):

    img = tf.convert_to_tensor(img)
    if img.dtype.is_integer:
        img = tf.cast(img, tf.float32) / 255.0
    else:
        img = tf.cast(img, tf.float32)
    return img

def show_lr_hr_pair_train(x_lr, x_hr, idx=None, method='bicubic', figsize=(9,3), title_prefix="Train"):


    if idx is None:
        idx = np.random.randint(0, len(x_lr))


    lr = _to_float01(x_lr[idx])
    hr = _to_float01(x_hr[idx])

    H, W = hr.shape[0], hr.shape[1]
    lr_up = tf.image.resize(lr, (H, W), method=method, antialias=True)


    lr_np     = np.clip(lr.numpy(),     0.0, 1.0)
    lr_up_np  = np.clip(lr_up.numpy(),  0.0, 1.0)
    hr_np     = np.clip(hr.numpy(),     0.0, 1.0)

    # ุฑุณู
    fig, axes = plt.subplots(1, 3, figsize=figsize)
    axes[0].imshow(lr_np)
    axes[0].set_title(f"{title_prefix} โข LR (small) #{idx}")
    axes[0].axis('off')

    axes[1].imshow(lr_up_np)
    axes[1].set_title("LR upsampled (bicubic)")
    axes[1].axis('off')

    axes[2].imshow(hr_np)
    axes[2].set_title("HR (ground truth)")
    axes[2].axis('off')

    plt.tight_layout()
    plt.show()



# TODO: ููุงุด ฺฉ ููููู ุงุฒ ุฏุงุฏูโูุง ุขููุฒุด

show_lr_hr_pair_train(x_train_lr, x_train, idx=0)
```

```python
# TODO: ุงุฌุงุฏ ุชุงุจุน Super Resolution

def conv_block(x, filters, k=3, act='relu', use_bn=False):
    x = layers.Conv2D(filters, k, padding='same', activation=act,
                      kernel_initializer='he_normal')(x)
    if use_bn:
        x = layers.BatchNormalization()(x)
    return x

def build_sr_model(
    lr_shape=(16,16,3),
    method='upsample',
    base_filters=64,
    num_blocks=3,
    use_bn=False
):

    assert method in {'upsample','transpose'}

    inputs = layers.Input(shape=lr_shape)


    x = conv_block(inputs, base_filters, use_bn=use_bn)


    for _ in range(num_blocks-1):
        x = conv_block(x, base_filters, use_bn=use_bn)


    if method == 'upsample':
        x = layers.UpSampling2D(size=(2,2), interpolation='bilinear')(x)
        x = conv_block(x, base_filters, use_bn=use_bn)
    else:  # method == 'transpose'
        x = layers.Conv2DTranspose(base_filters, kernel_size=4, strides=2,
                                   padding='same', activation='relu',
                                   kernel_initializer='he_normal')(x)


    x = conv_block(x, max(base_filters//2, 16), use_bn=use_bn)
    outputs = layers.Conv2D(3, 3, padding='same', activation='sigmoid')(x)

    model = models.Model(inputs, outputs, name=f"SR_CNN_x2_{method}")
    return model
```

```python
# TODO: ุจุง ุงุณุชูุงุฏู ุงุฒ ูุฏู ุขููุฒุดโุฏุฏูุ ุชุตุงูุฑ ุจุงฺฉูุช ุฑุง ุงุฒ ุฑู ุชุตุงูุฑ ุจโฺฉูุช ุชุณุช ูพุดโุจู ฺฉูุฏ.
# 8) ุงุฑุฒุงุจ ููุง ู ูุตูุฑุณุงุฒ ูุชุงุฌ

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt


X_test_lr = x_test_lr_np.astype("float32")
Y_test_hr = x_test_hr_np.astype("float32")


Y_pred_sr = model.predict(X_test_lr, batch_size=256, verbose=1)  # SR


def mean_psnr(y_true, y_pred, max_val=1.0):
    return float(tf.reduce_mean(tf.image.psnr(y_true, y_pred, max_val=max_val)).numpy())

def mean_ssim(y_true, y_pred, max_val=1.0):
    return float(tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=max_val)).numpy())


H, W = Y_test_hr.shape[1], Y_test_hr.shape[2]
Y_test_bic = tf.image.resize(X_test_lr, (H, W), method="bicubic", antialias=True).numpy()

print("Test โ Bicubic:",
      "PSNR:", mean_psnr(Y_test_hr, Y_test_bic),
      "SSIM:", mean_ssim(Y_test_hr, Y_test_bic))

print("Test โ Model  :",
      "PSNR:", mean_psnr(Y_test_hr, Y_pred_sr),
      "SSIM:", mean_ssim(Y_test_hr, Y_pred_sr))


def show_sr_triplets(x_lr, x_sr, x_hr, n=6, up_method="nearest", title="Test Triplets"):
    """
    x_lr: ุขุฑุงูโ LR ุจุง ุดฺฉู (N, h, w, 3)
    x_sr: ูพุดโุจู ูุฏู ุจุง ุดฺฉู (N, H, W, 3)
    x_hr: ูุฏู HR ุจุง ุดฺฉู (N, H, W, 3)
    n:   ุชุนุฏุงุฏ ูููููโูุง ุจุฑุง ููุงุด
    """
    N = len(x_lr)
    idxs = np.random.choice(N, size=min(n, N), replace=False)

    H, W = x_hr.shape[1], x_hr.shape[2]
    rows, cols = len(idxs), 3
    fig, axes = plt.subplots(rows, cols, figsize=(3.7*cols, 3.2*rows))
    if rows == 1:
        axes = np.expand_dims(axes, 0)

    for r, i in enumerate(idxs):

        lr_disp = tf.image.resize(x_lr[i], (H, W), method=up_method, antialias=True).numpy()
        panels = [lr_disp, x_sr[i], x_hr[i]]
        titles = ["LR (display โ)", "SR (predicted)", "HR (ground truth)"]

        for c in range(cols):
            ax = axes[r, c]
            ax.imshow(np.clip(panels[c], 0, 1))
            ax.set_title(titles[c])
            ax.axis("off")

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

show_sr_triplets(X_test_lr, Y_pred_sr, Y_test_hr, n=6, up_method="bicubic", title="Test Samples (LRโ vs SR vs HR)")

```

```python
# TODO: ูุญุงุณุจู PSNR ุจุฑุง ฺฉู ูุฌููุนู ุชุณุช
# ุงุณุชูุงุฏู ุงุฒ ุชุงุจุน psnr ุงุฒ TensorFlow ุจุฑุง ูุญุงุณุจู Peak Signal-to-Noise Ratio (PSNR)



import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt


X_test_lr = x_test_lr_np.astype("float32")
Y_test_hr = x_test_hr_np.astype("float32")


Y_pred_sr = model.predict(X_test_lr, batch_size=256, verbose=1)


H, W = Y_test_hr.shape[1], Y_test_hr.shape[2]
Y_test_bic = tf.image.resize(X_test_lr, (H, W), method="bicubic", antialias=True)


def psnr_vector(y_true, y_pred, max_val=1.0):

    y_true = tf.convert_to_tensor(y_true, dtype=tf.float32)
    y_pred = tf.convert_to_tensor(y_pred, dtype=tf.float32)
    return tf.image.psnr(y_true, y_pred, max_val=max_val).numpy()

def psnr_mean(y_true, y_pred, max_val=1.0):
    return float(np.mean(psnr_vector(y_true, y_pred, max_val=max_val)))


psnr_bic_vec  = psnr_vector(Y_test_hr, Y_test_bic)
psnr_sr_vec   = psnr_vector(Y_test_hr, Y_pred_sr)

print(f"PSNR (Test) โ Bicubic: {psnr_bic_vec.mean():.3f} dB")
print(f"PSNR (Test) โ Model  : {psnr_sr_vec.mean():.3f} dB")


```
